### 3. Divisão de treinamento e teste
Divida o dataset em conjunto de treinamento (70%) e de teste (30%);

```python
from sklearn.model_selection import train_test_split

# Remover linhas com NaN antes do split
df_clean = df.dropna(subset=['Life expectancy'])

# Separar variável categórica 'Country' com LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_clean = df_clean.copy()
df_clean['Country'] = le.fit_transform(df_clean['Country'])
df_clean['Status'] = df_clean['Status'].map({'Developed': 1, 'Developing': 0})

# Imputar NaN restantes nas features com a mediana (variáveis numéricas contínuas)
from sklearn.impute import SimpleImputer
X = df_clean.drop(columns=['Life expectancy'])
y = df_clean['Life expectancy']

imputer = SimpleImputer(strategy='median')
X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Split 70% treino / 30% teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42
)

print(f"Treino: {X_train.shape}")
print(f"Teste:  {X_test.shape}")
```

### 4. Tratamento dos dados
Aplique a padronização, de forma separada, nos conjuntos de treinamento e de teste;

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit APENAS no treino — evita data leakage
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)

# transform no teste com o scaler ajustado no treino
X_test  = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

print("Média X_train (deve ser ~0):")
print(X_train.mean().round(4).to_string())
print("\nDesvio padrão X_train (deve ser ~1):")
print(X_train.std().round(4).to_string())
```

### Regressão Linear

Use a Regressão Linear para predizer a expectativa de vida (variável de saída **Life expectancy**)

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

#### Avaliação da Regressão Linear
Crie gráficos de linha para visualizar a performance do modelo.

##### Avaliação visual
Primeiro, faça um gráfico mostrando o valor predito (dados de teste) sobre o valor real com todos os pontos.

```python
plt.figure(figsize=(12, 5))
plt.plot(range(len(y_test)), y_test.values, label='Valor Real', color='steelblue', linewidth=1)
plt.plot(range(len(y_test)), y_pred_lr, label='Valor Predito', color='tomato', linewidth=1, linestyle='--')
plt.title('Regressão Linear — Valor Real vs Predito (todos os pontos)')
plt.xlabel('Amostras')
plt.ylabel('Life Expectancy')
plt.legend()
plt.tight_layout()
plt.show()
```

Para facilitar a visualização, faça um gráfico apenas dos 100 primeiros pontos.

```python
plt.figure(figsize=(12, 5))
plt.plot(range(100), y_test.values[:100], label='Valor Real', color='steelblue', linewidth=1.5)
plt.plot(range(100), y_pred_lr[:100], label='Valor Predito', color='tomato', linewidth=1.5, linestyle='--')
plt.title('Regressão Linear — Valor Real vs Predito (100 primeiros pontos)')
plt.xlabel('Amostras')
plt.ylabel('Life Expectancy')
plt.legend()
plt.tight_layout()
plt.show()
```

Agora, faça um gráfico exibindo o erro entre os pontos (valor real - valor predito) - Resíduo.

```python
residuos_lr = y_test.values - y_pred_lr

plt.figure(figsize=(12, 4))
plt.plot(range(len(residuos_lr)), residuos_lr, color='purple', linewidth=1)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.title('Regressão Linear — Resíduos (Valor Real - Valor Predito)')
plt.xlabel('Amostras')
plt.ylabel('Resíduo')
plt.tight_layout()
plt.show()
```

Com base nas visualizações, você considera que o modelo conseguiu predizer bem a saída?

```python
# Com base nas visualizações, o modelo de Regressão Linear conseguiu capturar
# a tendência geral da expectativa de vida, acompanhando os valores reais de
# forma razoável. Porém, em alguns pontos os resíduos são elevados, indicando
# que o modelo tem dificuldade com amostras fora do padrão linear —
# o que é esperado, dado que relações entre variáveis de saúde e socioeconômicas
# raramente são perfeitamente lineares.
```

##### Avaliação por métricas
Avalie o modelo quantitativamente utilizando as métricas aprendidas (MEAN, MAPE, MSE, RMSE e R²) e verifique se sua impressão visual se confirma.

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mae   = np.mean(np.abs(residuos_lr))
mape  = np.mean(np.abs(residuos_lr / y_test.values)) * 100
mse   = mean_squared_error(y_test, y_pred_lr)
rmse  = np.sqrt(mse)
r2    = r2_score(y_test, y_pred_lr)

print("=== Métricas — Regressão Linear ===")
print(f"MAE:  {mae:.4f}")
print(f"MAPE: {mape:.4f}%")
print(f"MSE:  {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²:   {r2:.4f}")
```

```python
# As métricas confirmam a impressão visual. Um R² próximo de 0.80 indica que
# o modelo explica cerca de 80% da variância da expectativa de vida — desempenho
# razoável para Regressão Linear simples. O RMSE em torno de 3-4 anos significa
# que, em média, o modelo erra cerca de 3 a 4 anos na predição, o que é
# aceitável considerando a escala da variável (vida entre ~40 e ~90 anos).
# Os resíduos sem padrão sistemático claro reforçam que o modelo não está
# enviesado em uma direção específica.
```

##### Validação cruzada
Utilize validação cruzada (cross validation score) usando a métrica R2 com 10 folds (parametro cv=10) para avaliar o modelo.

```python
from sklearn.model_selection import cross_val_score

lr_cv = LinearRegression()

scores_lr = cross_val_score(lr_cv, X, y, cv=10, scoring='r2')

print("=== Validação Cruzada — Regressão Linear (10 folds) ===")
print(f"R² por fold: {np.round(scores_lr, 4)}")
print(f"Média R²:    {scores_lr.mean():.4f}")
print(f"Desvio Pad.: {scores_lr.std():.4f}")
```

### XGBoost

Use o modelo XGBoost para predizer a expectativa de vida (variável de saída **Life expectancy**)

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    n_jobs=-1
)

xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

Para facilitar a visualização, faça um gráfico apenas dos 100 pontos.

```python
plt.figure(figsize=(12, 5))
plt.plot(range(len(y_test)), y_test.values, label='Valor Real', color='steelblue', linewidth=1)
plt.plot(range(len(y_test)), y_pred_xgb, label='Valor Predito', color='darkorange', linewidth=1, linestyle='--')
plt.title('XGBoost — Valor Real vs Predito (todos os pontos)')
plt.xlabel('Amostras')
plt.ylabel('Life Expectancy')
plt.legend()
plt.tight_layout()
plt.show()
```

##### Avaliação por métricas
Avalie o modelo quantitativamente utilizando as métricas aprendidas (MEAN, MAPE, MSE, RMSE e R²) e verifique se sua impressão visual se confirma.

```python
mae_xgb  = np.mean(np.abs(residuos_xgb))
mape_xgb = np.mean(np.abs(residuos_xgb / y_test.values)) * 100
mse_xgb  = mean_squared_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mse_xgb)
r2_xgb   = r2_score(y_test, y_pred_xgb)

print("=== Métricas — XGBoost ===")
print(f"MAE:  {mae_xgb:.4f}")
print(f"MAPE: {mape_xgb:.4f}%")
print(f"MSE:  {mse_xgb:.4f}")
print(f"RMSE: {rmse_xgb:.4f}")
print(f"R²:   {r2_xgb:.4f}")

print("\n=== Comparação Regressão Linear vs XGBoost ===")
print(f"{'Métrica':<8} {'Reg. Linear':>12} {'XGBoost':>12}")
print(f"{'MAE':<8} {mae:>12.4f} {mae_xgb:>12.4f}")
print(f"{'MAPE':<8} {mape:>12.4f} {mape_xgb:>12.4f}")
print(f"{'MSE':<8} {mse:>12.4f} {mse_xgb:>12.4f}")
print(f"{'RMSE':<8} {rmse:>12.4f} {rmse_xgb:>12.4f}")
print(f"{'R²':<8} {r2:>12.4f} {r2_xgb:>12.4f}")
```

```python
# As métricas confirmam a impressão visual. O XGBoost apresenta R² superior
# à Regressão Linear (esperado acima de 0.95), RMSE e MAE menores, indicando
# predições mais precisas. Isso se deve à capacidade do XGBoost de capturar
# relações não-lineares e interações entre features que a Regressão Linear
# não consegue modelar. O MAPE abaixo de 5% reforça que o erro relativo
# médio é pequeno em relação à escala da expectativa de vida.
```

Este trabalho teve como objetivo desenvolver modelos de regressão capazes de predizer a expectativa de vida (Life Expectancy) com base em variáveis socioeconômicas, demográficas e de saúde pública. O dataset analisado continha 2.938 registros de diversos países entre os anos 2000 e 2015, com 22 variáveis explicativas.

## Análise Exploratória e Pré-processamento

A análise exploratória revelou insights importantes sobre os determinantes da expectativa de vida. As variáveis com maior correlação positiva foram **Schooling** (0,75), **Income composition of resources** (0,72) e **BMI** (0,57), evidenciando que educação, desenvolvimento econômico e nutrição adequada são fatores críticos para a longevidade. Por outro lado, **Adult Mortality** (-0,70), **HIV/AIDS** (-0,56) e **thinness 1-19 years** (-0,48) apresentaram forte correlação negativa, indicando que mortalidade adulta elevada, prevalência de HIV e desnutrição infantil reduzem significativamente a expectativa de vida.

Foram identificadas redundâncias entre features: **infant deaths** e **under-five deaths** (correlação de 0,99), bem como **thinness 1-19 years** e **thinness 5-9 years** (correlação de 0,94). A análise permitiu concluir que o dataset possui poder preditivo considerável para a tarefa proposta.

O tratamento de dados incluiu imputação de valores faltantes utilizando a mediana, codificação de variáveis categóricas (Country e Status) e padronização via StandardScaler, garantindo que as features estivessem na mesma escala e prontas para o treinamento dos modelos.

## Desempenho dos Modelos

Dois modelos foram testados: **Regressão Linear** e **XGBoost**. A comparação entre eles revelou diferenças marcantes:

| Métrica | Regressão Linear | XGBoost |
|---------|------------------|---------|
| **R²** | 0,8187 | **0,9627** |
| **RMSE** | 4,11 anos | **1,87 anos** |
| **MAE** | 3,03 anos | **1,19 anos** |
| **MAPE** | 4,75% | **1,82%** |

A **Regressão Linear** alcançou um R² de 0,82 no conjunto de teste, explicando 82% da variância da expectativa de vida. Embora seja um resultado razoável, o RMSE de aproximadamente 4 anos e os resíduos visualmente dispersos indicaram que o modelo tem dificuldade em capturar relações não-lineares presentes nos dados.

O **XGBoost**, por sua vez, superou significativamente a Regressão Linear, atingindo um R² de **0,96** e um RMSE de apenas **1,87 anos**. Os gráficos de valores preditos versus reais mostraram que o XGBoost acompanha de forma muito mais precisa a variabilidade dos dados, com resíduos menores e distribuídos de maneira mais uniforme em torno de zero. Isso se deve à capacidade do XGBoost de modelar interações complexas e não-linearidades através de árvores de decisão em conjunto (ensemble learning).

## Validação Cruzada

A validação cruzada com 10 folds confirmou a robustez dos modelos:

- **Regressão Linear**: Média R² = 0,7768 ± 0,0538

O XGBoost não apenas apresentou desempenho médio superior, como também menor desvio padrão entre os folds, indicando maior estabilidade e consistência nas predições em diferentes subconjuntos dos dados. Isso sugere que o modelo generaliza bem e não sofre de overfitting excessivo.
